Function to optimize:

```python
def f(x, mu, cov_matrix, q=0.1):    
    x = np.asarray(x)
    mu = np.asarray(mu)
    cov_matrix = np.asarray(cov_matrix)

    benefit = np.dot(x, mu)
    risk = np.einsum('i,ij,j->', x, cov_matrix, x)

    return benefit - q * risk

f(x, mu, cov_matrix)
```

We can use qiskit-optimization to build a qubo problem

```python
import numpy as np
from qiskit_optimization import QuadraticProgram

def markowitz_qubo(mu, cov_matrix, q=0.1, max_assets=5):
    num_assets = len(mu)
    
    # Create a Quadratic Program
    qubo = QuadraticProgram()
    
    # Define binary variables x_i (selecting an asset or not)
    for i in range(num_assets):
        qubo.binary_var(f"x{i}")
    
    # Define the objective function: Minimize (- benefit + q * risk)
    linear_coeffs = -mu  # Corresponds to -x^T mu
    quadratic_coeffs = q * cov_matrix  # Corresponds to q x^T Î£ x
    
    # Add to Qiskit's QuadraticProgram
    qubo.maximize(linear=linear_coeffs, quadratic=quadratic_coeffs)
    qubo.linear_constraint([1] * num_assets, "<=", max_assets, "max_assets_constraint")
    
    return qubo

# Create QUBO
qubo_problem = markowitz_qubo(np.asarray(mu), np.asarray(cov_matrix))

# Print QUBO formulation
print(qubo_problem.prettyprint())
```

```python
import numpy as np
from qiskit_optimization import QuadraticProgram
from qiskit_optimization.algorithms import MinimumEigenOptimizer
from qiskit_algorithms import QAOA
from qiskit.primitives import Sampler

def create_qubo_portfolio(cov_matrix, expected_returns, risk_weight=1, return_weight=1):
    """
    Crea un problema QUBO para la optimizaciÃ³n de cartera.
    
    :param cov_matrix: Matriz de covarianza de los retornos
    :param expected_returns: Vector de retornos esperados
    :param risk_weight: Peso para el factor de riesgo
    :param return_weight: Peso para el factor de retorno
    :return: Problema QUBO formulado
    """
    n = len(expected_returns)
    
    # Crear el programa cuadrÃ¡tico
    qp = QuadraticProgram('Portfolio Optimization')
    
    # AÃ±adir variables binarias
    for i in range(n):
        qp.binary_var(name=f'x{i}')
    
    # AÃ±adir funciÃ³n objetivo
    linear = {}
    quadratic = {}
    
    # TÃ©rminos lineales (retornos esperados)
    for i in range(n):
        linear[f'x{i}'] = -return_weight * expected_returns[i]
    
    # TÃ©rminos cuadrÃ¡ticos (matriz de covarianza)
    for i in range(n):
        for j in range(i, n):
            if i == j:
                quadratic[(f'x{i}', f'x{j}')] = risk_weight * cov_matrix[i][j]
            else:
                quadratic[(f'x{i}', f'x{j}')] = 2 * risk_weight * cov_matrix[i][j]
    
    qp.minimize(linear=linear, quadratic=quadratic)
    
    return qp

# Ejemplo de uso
if __name__ == "__main__":
    # Ejemplo de matriz de covarianza (3x3)
    # cov_matrix = np.array([
    #     [0.04, 0.02, 0.01],
    #     [0.02, 0.09, 0.03],
    #     [0.01, 0.03, 0.16]
    # ])
    cov_matrix = cov_matrix.values
    
    # Ejemplo de retornos esperados
    expected_returns = np.array([0.10, 0.15, 0.20])
    
    # Crear el problema QUBO
    qubo = create_qubo_portfolio(cov_matrix, expected_returns)
    
    print("Problema QUBO formulado:")
    print(qubo.prettyprint())
    
    # Resolver el problema usando QAOA
    # # optimizer = SPSA(maxiter=100)
    # optimizer = None
    # qaoa = QAOA(sampler=Sampler(), optimizer=optimizer)
    # optimizer = MinimumEigenOptimizer(qaoa)
    # result = optimizer.solve(qubo_problem)
    
    # print("\nSoluciÃ³n Ã³ptima:")
    # print(result.x)
    
    # print("\nValor de la funciÃ³n objetivo:")
    # print(result.fval)
```
